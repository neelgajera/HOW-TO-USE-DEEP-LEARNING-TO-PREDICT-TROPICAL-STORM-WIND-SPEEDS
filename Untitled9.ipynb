```
from tqdm import tqdm
import requests
 
chunk_size = 1024
 
url = "https://drivendata-public-assets.s3.amazonaws.com/re-train-images.tgz"
 
req = requests.get(url, stream = True)
 
total_size = int(req.headers['content-length'])
 
with open("p.tgz", "wb") as file:
    for data in tqdm(iterable=req.iter_content(chunk_size=chunk_size), total = total_size/chunk_size, unit='KB'):
        file.write(data)
```

```
import os, sys, tarfile
def extract(tar_url, extract_path='.'):
    
    tar = tarfile.open(tar_url, 'r')
    for item in tar:
        tar.extract(item, extract_path)
        if item.name.find(".tgz") != -1 or item.name.find(".tar") != -1:
            extract(item.name, "./" + item.name[:item.name.rfind('/')])
```

```
def load_image(path):
    # Read image from path.
    image = tf.io.read_file(path)
    # Decode the jpeg image to array [0, 255].

    image = tf.image.decode_jpeg(image)
    
    return image
def parse_records(line):
    image = load_image(line)
    return image
  
```

```

data_imagep = tf.data.Dataset.from_tensor_slices(train.file_name)
data_img = data_imagep.map(parse_records)
x_train = train.drop(["wind_speed","image_id","storm_id","file_name"],axis=1)

dataset_12 = tf.data.Dataset.from_tensor_slices(x_train)

dataset_label = tf.data.Dataset.from_tensor_slices(train.wind_speed)

dataset_f = tf.data.Dataset.zip((data_img, dataset_12))
dataset = tf.data.Dataset.zip((dataset_f, dataset_label))
```

```
base_model = tf.keras.applications.InceptionResNetV2(
    include_top=False, weights='imagenet', input_tensor=None,
    input_shape=(366,366,3), pooling='avg'
)
```

```
##base_model.trainable = False
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_dataset =dataset.skip(2000)
vaild_dataset = dataset.take(2000)
train_dataset = train_dataset.cache().prefetch(buffer_size=AUTOTUNE)
vaild_dataset = vaild_dataset.cache().prefetch(buffer_size=AUTOTUNE)
```

```
inputs = tf.keras.Input(shape=(366,366,1), name ="q")
x = tf.keras.layers.Conv2D(3,(3,3),padding='same')(inputs)
x = tf.keras.applications.inception_resnet_v2.preprocess_input(x)
x = base_model(x)
x = tf.keras.layers.Dropout(0.2)(x) 
x = tf.keras.layers.Dense(500)(x)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.Model(inputs=inputs, outputs=x)
y = tf.keras.Sequential([
  tf.keras.layers.Dense(2, activation=tf.nn.relu, input_shape=(2,1)),
  tf.keras.layers.Dense(10, activation=tf.nn.relu),
  tf.keras.layers.Dense(2),
  tf.keras.layers.Flatten()
])


z = tf.keras.layers.concatenate(inputs=[x.output, y.output],axis=1)
print(z)
z = tf.keras.layers.Dense(300)(z)
z = tf.keras.layers.Dropout(0.2)(z) # Regularize with dropout
outputs = tf.keras.layers.Dense(1)(z)
model = tf.keras.Model([x.input,y.input],outputs)
```

```
checkpoint_filepath = '/content/drive/MyDrive'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_best_only=True)
```
